```python
import subprocess
subprocess.run(["pip", "install", "requests"], capture_output=True)
print("requests library ready!")
import requests

# --- Step 1: Fetch 20 products ---
url = "https://dummyjson.com/products?limit=20"
response = requests.get(url, timeout=5)
data = response.json()

print(f"Status Code: {response.status_code}")
print(f"Total products fetched: {len(data['products'])}")
print(f"First product: {data['products'][0]}")
# --- Print formatted table ---
print(f"\n{'ID':<4} | {'Title':<30} | {'Category':<13} | {'Price':<8} | {'Rating'}")
print("-" * 70)

for product in data["products"]:
    pid      = product["id"]
    title    = product["title"][:28]  # trim long titles
    category = product["category"]
    price    = f"${product['price']:.2f}"
    rating   = product["rating"]

    print(f"{pid:<4} | {title:<30} | {category:<13} | {price:<8} | {rating}")
    # --- Step 2: Filter products with rating >= 4.5 and sort by price descending ---
filtered = [p for p in data["products"] if p["rating"] >= 4.5]
sorted_products = sorted(filtered, key=lambda x: x["price"], reverse=True)

print(f"\nProducts with rating >= 4.5 (sorted by price high to low):")
print(f"\n{'ID':<4} | {'Title':<30} | {'Category':<13} | {'Price':<8} | {'Rating'}")
print("-" * 70)

for product in sorted_products:
    pid      = product["id"]
    title    = product["title"][:28]
    category = product["category"]
    price    = f"${product['price']:.2f}"
    rating   = product["rating"]
    print(f"{pid:<4} | {title:<30} | {category:<13} | {price:<8} | {rating}")

    # --- Step 3: Fetch laptops category ---
laptop_url = "https://dummyjson.com/products/category/laptops"
laptop_response = requests.get(laptop_url, timeout=5)
laptop_data = laptop_response.json()

print(f"\nLaptops available ({len(laptop_data['products'])} found):")
print("-" * 40)
for laptop in laptop_data["products"]:
    print(f"  {laptop['title']:<35} ${laptop['price']:.2f}")

    # --- Step 4: POST request to add a new product ---
post_url = "https://dummyjson.com/products/add"
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

post_response = requests.post(post_url, json=new_product, timeout=5)
post_data = post_response.json()

print("\nPOST Request Response:")
print("-" * 40)
print(f"  ID          : {post_data['id']}")
print(f"  Title       : {post_data['title']}")
print(f"  Price       : {post_data['price']}")
print(f"  Category    : {post_data['category']}")
print(f"  Description : {post_data['description']}")
```

```python
# ============================================================
# TASK 3A - Guarded Calculator
# ============================================================

def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

# Test all three cases
print(safe_divide(10, 2))      # normal division
print(safe_divide(10, 0))      # division by zero
print(safe_divide("ten", 2))   # invalid types
# ============================================================
# TASK 3B - Guarded File Reader
# ============================================================

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
            return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

# Test with existing file
print("Testing with python_notes.txt:")
print("-" * 40)
result = read_file_safe("python_notes.txt")
if result:
    print(result)

# Test with non-existing file
print("Testing with ghost_file.txt:")
print("-" * 40)
read_file_safe("ghost_file.txt")

# ============================================================
# TASK 3C - Robust API Calls
# ============================================================

def fetch_products():
    try:
        response = requests.get("https://dummyjson.com/products?limit=20", timeout=5)
        data = response.json()
        print(f"✅ Products fetched successfully! Total: {len(data['products'])}")
        return data["products"]
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")

# Test it
fetch_products()

# ============================================================
# TASK 3D - Input Validation Loop
# ============================================================

while True:
    user_input = input("Enter a product ID to look up (1-100), or 'quit' to exit: ")

    # Check if user wants to quit
    if user_input == "quit":
        print("Goodbye!")
        break

    # Validate input is a number
    try:
        product_id = int(user_input)
    except ValueError:
        print("⚠️ Please enter a valid number between 1 and 100!")
        continue

    # Validate range
    if product_id < 1 or product_id > 100:
        print("⚠️ Product ID must be between 1 and 100!")
        continue

    # Make API call with valid input
    try:
        response = requests.get(f"https://dummyjson.com/products/{product_id}", timeout=5)
        if response.status_code == 200:
            product = response.json()
            print(f"✅ Product found: {product['title']} — ${product['price']}")
        elif response.status_code == 404:
            print("Product not found.")
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
```

```python
# ============================================================
# TASK 4 - Logging to File
# ============================================================

from datetime import datetime

def log_error(function_name, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"

    with open("/content/error_log.txt", "a", encoding="utf-8") as f:
        f.write(log_entry)

    print(f"📝 Logged: {log_entry.strip()}")

# Test the logger works
log_error("test_function", "TestError", "This is a test log entry")

# --- Trigger 1: ConnectionError with unreachable URL ---
try:
    response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    print("Connection failed. Please check your internet.")
    log_error("fetch_products", "ConnectionError", "No connection could be made")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
    log_error("fetch_products", "Timeout", "Request timed out")

# --- Trigger 2: HTTP 404 error with invalid product ID ---
response = requests.get("https://dummyjson.com/products/999", timeout=5)
if response.status_code != 200:
    log_error("lookup_product", "HTTPError", f"404 Not Found for product ID 999")
    print("Product not found.")

# --- Print full error log ---
print("\n--- Contents of error_log.txt ---")
with open("/content/error_log.txt", "r", encoding="utf-8") as f:
    print(f.read())
```
