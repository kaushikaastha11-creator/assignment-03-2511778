```python
# ============================================================
# TASK 1 - Data Parsing & Profile Cleaning
# ============================================================

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

for student in raw_students:
    name = student["name"].strip().title()
    roll = int(student["roll"])
    marks = [int(x) for x in student["marks_str"].split(", ")]

    # Validate name - every word must contain only letters
    words = name.split()
    is_valid = True
    for word in words:
        if not word.isalpha():
            is_valid = False
            break

    if is_valid:
        print(f"{name} ✓ Valid name")
    else:
        print(f"{name} ✗ Invalid name")

    # Profile card
    print("================================")
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("================================")

    # Print ALL CAPS and lowercase for roll 103
    if roll == 103:
        print(f"ALL CAPS  : {name.upper()}")
        print(f"lowercase : {name.lower()}")
```

```python
# ============================================================
# TASK 2 - Marks Analysis Using Loops & Conditionals
# ============================================================

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# --- For loop: print each subject, marks and grade ---
for subject, mark in zip(subjects, marks):
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject} : {mark} : {grade}")

# --- Calculations ---
total = sum(marks)
average = round(total / len(marks), 2)

print(f"\nTotal Marks   : {total}")
print(f"Average Marks : {average}")

# Highest scoring subject
highest_mark = max(marks)
highest_index = marks.index(highest_mark)
print(f"Highest Subject : {subjects[highest_index]} : {highest_mark}")

# Lowest scoring subject
lowest_mark = min(marks)
lowest_index = marks.index(lowest_mark)
print(f"Lowest Subject : {subjects[lowest_index]} : {lowest_mark}")

# --- While loop: add new subjects ---
count = 0

while True:
    subject_input = input("Enter subject name (or 'done' to stop): ")

    if subject_input == "done":
        break

    marks_input = input(f"Enter marks for {subject_input}: ")

    try:
        m = int(marks_input)
        if 0 <= m <= 100:
            subjects.append(subject_input)
            marks.append(m)
            count += 1
        else:
            print("⚠️ Marks must be between 0 and 100. Subject not added.")
    except ValueError:
        print("⚠️ Invalid input, please enter a number. Subject not added.")

print(f"\nNew subjects added : {count}")
print(f"Updated average    : {round(sum(marks) / len(marks), 2)}")
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print(f"{'Name':<18}| {'Average':<7} | {'Status'}")
print("-" * 40)

for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    status = "Pass" if avg >= 60 else "Fail"
    print(f"{name:<18}| {avg:<7} | {status}")
    # Count pass and fail
passed = sum(1 for name, marks in class_data if round(sum(marks)/len(marks), 2) >= 60)
failed = len(class_data) - passed
print(f"\nPassed : {passed} | Failed : {failed}")

# Class topper
topper = max(class_data, key=lambda x: round(sum(x[1])/len(x[1]), 2))
topper_avg = round(sum(topper[1])/len(topper[1]), 2)
print(f"Class Topper : {topper[0]} : {topper_avg}")

# Class average
class_avg = round(sum(round(sum(marks)/len(marks), 2) for name, marks in class_data) / len(class_data), 2)
print(f"Class Average : {class_avg}")
```

```python
# ============================================================
# TASK 4 - String Manipulation Utility
# ============================================================

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1 - Strip whitespace
clean_essay = essay.strip()
print("Step 1 - Stripped:")
print(clean_essay)
# Step 2 - Title Case
print("\nStep 2 - Title Case:")
print(clean_essay.title())

# Step 3 - Count "python"
print("\nStep 3 - Count of 'python':")
print(clean_essay.count("python"))
# Step 4 - Replace "python" with "Python 🐍"
print("\nStep 4 - Replaced:")
print(clean_essay.replace("python", "Python 🐍"))

# Step 5 - Split into sentences
sentences = clean_essay.split(". ")
print("\nStep 5 - Sentences list:")
print(sentences)

# Step 6 - Numbered sentences
print("\nStep 6 - Numbered sentences:")
for i, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence = sentence + "."
    print(f"{i}. {sentence}")
```
